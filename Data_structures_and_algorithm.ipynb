````python
from math import gcd
from functools import reduce

def find_greatest_divisor_with_same_remainder(nums):
    if len(nums) < 2:
        return None  # need at least 2 numbers

    # Take difference of each element with the first
    diffs = [abs(nums[i] - nums[0]) for i in range(1, len(nums))]

    # Compute GCD of all differences
    return reduce(gcd, diffs)

# Example usage
nums = [43,91,183]
result = find_greatest_divisor_with_same_remainder(nums)
print("Greatest number that divides all and leaves same remainder:", result)
````

````python
#Patterns
''' ****
    ****
    ****
    ****'''
for i in range(4):
  for j in range(4):
    print('*', end = "")
  print()

'''
  *
  **
  ***
  ****
  *****
  ******
  '''
print("---------------------" ,end='')
for i in range(0,8):
  for j in range(i):
      print('*', end = "")
  print()
'''
1
12
123
1234
12345
123456
1234567
12345678
'''

print("---------------------" ,end='')
for i in range(10):
  for j in range(1,i):
      print(j, end = "")
  print()

print("---------------------" ,end='')


'''
1
22
333
4444
55555'''
for i in range(0,6):
  for j in range(i):
    print(i, end = "")
  print()
print("---------------------" ,end='\n')
def reversed_stars(n):
  for i in range (n,0,-1):
    for j in range(i):
      print('*', end = "")
    print()
reversed_stars(4)


'''
     *
   * * *
 * * * * *
 '''
print("---------------------" ,end='\n')
n =4
for i in range(n):
  for j in range(n-i-1):
    print(" ", end="")
  for k in range(2* i +1):
    print("*" ,end ="")
  for j in range(n-i-1):
    print(" ", end="")
  print()


print("---------------------" ,end='\n')
n =4
for i in range(n):
  for j in range(i):
    print(" ", end="")
  for k in range(2*n -(2* i +1)):
    print("*" ,end ="")
  for j in range(i):
    print(" ", end="")
  print()
print("---------------------" ,end='\n')
n =9
for i in range(1,2*n-1):
  stars = i
  if (i>n):
    stars = 2*n -i
  for j in range(1,stars ):
    print("*", end="")
  print()


print("---------------------" ,end='\n')

'''
1
0 1
1 0 1
0 1 0 1
'''
n = 10
start = 1
for i in range(1,n):
  if i%2==0:
    start =1
  else:
    start =0
  for j in range(1,i):
    print(str(start) + " ", end="")
    start = 1-start
  print()
print("---------------------" ,end='\n')
n =5
space = 2*(n-2)
for i in range(1,n+1):
  for j in range(i+1):
    print(j, end='')
  for k in range(space):
    print(" ", end='')
  for l in range(i,1,-1):
    print(l, end='')
  print()
  space -= -2
print("---------------------" ,end='\n')
num = 1
for i in range(n):
  for j in range(i):
    print(num, end=' ')
    num +=1
  print()

print("---------------------", end='\n')
'''
A
A B
A B C
A B C D'''
for i in range(n):
  for j in range(65 , 65+i):
    print(chr(j), end=' ')
  print()
print("---------------------", end='\n')
for i in range(n,0,-1):
  for j in range(65 , 65+i):
    print(chr(j), end=' ')
  print()
print("---------------------", end='\n')
for i in range(0,n):
  char = chr(ord('A')+i)
  for j in range(i+1):
    print(char, end=' ')
  print()
print("---------------------", end='\n')

print("---------------------" ,end='\n')
n =4
for i in range(n):
  for j in range(n-i-1):
    print(" ", end="")
  char = ord('A')
  for k in range(2 * i + 1):
      print(chr(char), end=" ")
      if k < i:
          char += 1
      else:
          char -= 1
  for j in range(n-i-1):
    print(" ", end="")
  print()
print("---------------------", end='\n')
n=5
start = ord('A') + n - 1

for i in range(n):
    for j in range(start, start - i - 1, -1):
        print(chr(j), end='')
    print()

print(">>> COUNT DIGIT <<<")
n =0
count = 0
if n ==0 :
  count =1
while n!=0:
  n = n//10
  count +=1
print(count)
print("---------------------", end='\n')
print("ARMSTRONG NUMBER", end='\n')
def countnumberofdigit(n):
  if n ==0 :
    return 1
  count = 0
  while n!=0:
    n = n//10
    count +=1
  return count
def check_armstrong_number(n):
  temp = 0
  if n ==0 :
    return 1
  count = countnumberofdigit(n)
  temp = n
  sum_digits = 0
  while temp != 0:
      digit = temp % 10
      print("digit =", digit)
      sum_digits += digit ** count
      temp = temp // 10
      print("partial sum =", sum_digits)
      print("Final sum =", sum_digits)
  if sum_digits == n:
      print(f"{n} is an Armstrong number ðŸ’ª")
  else:
      print(f"{n} is NOT an Armstrong number âŒ")
check_armstrong_number(153)


print("Find all divisor of a number ")



import math

def findDivisors(n):
    # Initialize an empty
    # list to store the divisors
    divisors = []

    # Iterate up to the square
    # root of n to find divisors
    # Calculate the square root of n
    sqrtN = int(math.sqrt(n))

    # Loop from 1 to the
    # square root of n
    for i in range(1, sqrtN + 1):
        # Check if i divides n
        # without leaving a remainder
        if n % i == 0:
            # Add divisor i to the list
            divisors.append(i)

            # Add the counterpart divisor
            # if it's different from i
            if i != n // i:
                # Add the counterpart
                # divisor to the list
                divisors.append(n // i)

    # Return the list of divisors
    return divisors

if __name__ == "__main__":
    number = 12
    divisors = findDivisors(number)

    print("Divisors of", number, "are:", end=" ")
    for divisor in divisors:
        print(divisor, end=" ")
    print()


````

````python
## brute force only works for smaller arrays
def maxFrequency(nums, k) :
        hashmap = {}
        max_ele = max(nums)
        temp = 0
        for i in range(len(nums)):
          if nums[i] < max_ele:
            for j in range(k):
              nums[i]+=1
              if nums[i] >= max_ele:
                break
          if nums[i] in hashmap:
            hashmap[nums[i]] +=1
          else:
            hashmap[nums[i]]=1
        for k,v in hashmap.items():
          if v > temp :
            temp = v
        return v
print("using brute force (n**2): ",maxFrequency([1,2,4,],5))
'''
ðŸ’¡ Idea:
Weâ€™ll use a sliding window, always trying to make all values in the window equal to the rightmost value (nums[r]).

We keep a total = sum of all elements in window [l...r].

We want:

nums[r] * (window size) <= total + k
If this is true, we can convert all numbers to nums[r] using at most k operations.

ðŸ§® Step-by-Step:
Step 1:
r = 0, nums[r] = 1

Window: [1]

total = 1

Check: 1 * 1 <= 1 + 5 â†’ âœ…
â†’ res = 1

Step 2:
r = 1, nums[r] = 2

Window: [1, 2]

total = 1 + 2 = 3

Check: 2 * 2 = 4 <= 3 + 5 = 8 â†’ âœ…
â†’ res = 2

Step 3:
r = 2, nums[r] = 4

Window: [1, 2, 4]

total = 1 + 2 + 4 = 7

Check: 4 * 3 = 12 <= 7 + 5 = 12 â†’ âœ…
â†’ res = 3

âœ… Final Answer: 3
Meaning â†’ You can make 3 numbers equal with at most 5 increments.
'''
def maxFrequency(nums: list[int], k: int) -> int:
    nums.sort()
    l = 0
    total = 0
    res = 0

    for r in range(len(nums)):
        total += nums[r]

        # Check if total + k >= nums[r] * window_size
        while nums[r] * (r - l + 1) > total + k:
            total -= nums[l]
            l += 1

        res = max(res, r - l + 1)

    return res
print("using sliding window (nlogn ): ",maxFrequency([1,2,4,],k=5))
````

**Merge sort**


ðŸ§  â€“ Concept:
Divide the array recursively into halves until each part has only 1 element.

Then merge them back in sorted order.

Weâ€™ll use this array:

arr = [3, 44, 23, 12, 4, 5]

**Initial Array:**
$$
3 \quad 44 \quad 23 \quad 12 \quad 4 \quad 5
$$

---

### ðŸ”» Divide Phase:

1. Divide into 2 halves:
$$
[3, 44, 23] \quad\quad [12, 4, 5]
$$

2. Further divide:
$$
[3] \quad [44, 23] \quad\quad [12] \quad [4, 5]
$$

3. Further:
$$
[3] \quad [44] \quad [23] \quad\quad [12] \quad [4] \quad [5]
$$

---

### ðŸ”¼ Merge Phase:

#### ðŸ”¸ Merge [44] and [23] â†’ [23, 44]
$$
[3] \quad
\begin{array}{|c|c|}
\hline
23 & 44 \\
\hline
\end{array}
\quad\quad [12] \quad [4] \quad [5]
$$

#### ðŸ”¸ Merge [3] and [23, 44] â†’ [3, 23, 44]
$$
\begin{array}{|c|c|c|}
\hline
3 & 23 & 44 \\
\hline
\end{array}
\quad\quad [12] \quad [4] \quad [5]
$$

#### ðŸ”¸ Merge [4] and [5] â†’ [4, 5]
$$
[12] \quad
\begin{array}{|c|c|}
\hline
4 & 5 \\
\hline
\end{array}
$$

#### ðŸ”¸ Merge [12] and [4, 5] â†’ [4, 5, 12]
$$
\begin{array}{|c|c|c|}
\hline
4 & 5 & 12 \\
\hline
\end{array}
$$

#### ðŸ”¸ Final Merge [3, 23, 44] and [4, 5, 12]:
**Step 1: 3 < 4** â†’ 3
**Step 2: 4 < 23** â†’ 4  
**Step 3: 5 < 23** â†’ 5  
**Step 4: 12 < 23** â†’ 12  
**Step 5: 23  
Step 6: 44**

$$
\begin{array}{|c|c|c|c|c|c|}
\hline
3 & 4 & 5 & 12 & 23 & 44 \\
\hline
\end{array}
$$

---

### âœ… Final Sorted Array:


ðŸ•’ Time Complexity of Merge Sort:

Case	Time
Best	O(n log n)
Average	O(n log n)
Worst	O(n log n)
Because it divides the array into halves â‡’ log n

And merges all n elements at each level â‡’ O(n)

So total â‡’ O(n log n)

ðŸ’¾ Space Complexity:
O(n) extra space required (for temporary arrays during merge)

ðŸ”¥ Merge Sort is not in-place due to the extra space.

````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Split array into two halves
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    # Merge the sorted halves
    return merge(left, right)

def merge(left, right):
    merged = []
    i = j = 0

    # Compare and merge elements
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Add remaining elements
    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged

# Example usage:
arr = [5, 2, 8, 1, 3]
sorted_arr = merge_sort(arr)
print("Sorted:", sorted_arr)
````
